```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
ax.plot([1, 2, 3, 4], [100, 50, 75, -2], linewidth=1, color='blue', label='foo')
ax.plot([1, 2, 3, 4], [10, 50, 20, 70], linewidth=2, color='black', linestyle='--', label='bar')
ax.plot([1, 2, 3, 4], [10, 20, 30, 80], linewidth=10, color='green', linestyle=':', alpha=.5, label='foobar')
ax.legend()
ax.set_xlabel('pigs')
ax.set_ylabel('pigeons')
ax.set_title('pigs vs pigeons')
ax.set_xlim((1, 10))
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(np.random.random(100), np.random.random(100), marker='^',
            c=np.random.random(100), cmap='bwr')
ax.plot(np.linspace(0, 1, 20), np.random.random(20), color='green')
```

```python
fig, axs = plt.subplots(2, 3, figsize=(12, 8))
axs[0][0].plot(np.linspace(0, 1, 20), np.random.random(20))
axs[0][0].set_title('oops')
axs[0][1].scatter(np.linspace(0, 1, 20), np.random.random(20))
axs[0][1].set_title('more oops')
axs[0][2].scatter(np.linspace(0, 1, 20), np.random.random(20), color='yellow')
axs[1][0].plot(np.linspace(0, 1, 20), np.random.random(20), color='red')
axs[1][1].scatter(np.linspace(0, 1, 20), np.random.random(20), color='purple')
axs[1][2].bar(np.linspace(0, 1, 3), np.random.random(3), color='blue')
fig.suptitle('My drawings')
```

```python
fig, ax = plt.subplots()
ax.bar([1, 2, 3, 4], [4, 3, 2, 5])
ax.set_xticks([1, 2, 3, 4])
ax.set_xticklabels(['hello', 'world', 'whats', 'up'], rotation=45)
plt.show()
```

```python
school1_heights = np.random.random(size=100) + .5
school2_heights = np.random.random(size=100) + .5
```

```python
fig, ax = plt.subplots()
ax.hist([school1_heights, school2_heights], label=['school 1', 'school 2'])
ax.set_xlabel('Height [m]')
ax.set_ylabel('# children')
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots()
ax.pie([1, 2, 3, 4], labels=['bread', 'fruit', 'vegetable', 'pasta'], explode=[.2, .8, 0, 0],
      autopct='%d%%')
plt.show()
```
