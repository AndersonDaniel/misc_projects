```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = sns.load_dataset('diamonds')
```

```python
df.shape
```

```python
df.head()
```

```python
sns.pairplot(data=df, vars=['carat', 'depth', 'table', 'price', 'x', 'y', 'z'])
```

```python
sns.countplot(df['cut'])
```

```python
sns.countplot(df['color'])
```

```python
sns.countplot(df['clarity'])
```

```python
sns.violinplot(data=df, x='cut', y='price')
```

```python
sns.distplot(df['price'], hist=False)
```

```python
sns.distplot(np.log1p(df['price']), hist=False)
```

```python
df['price_log1p'] = np.log1p(df['price'])
```

```python
sns.violinplot(data=df, x='cut', y='price_log1p')
```

```python
sns.violinplot(data=df, x='cut', y='carat')
```

```python
sns.distplot(df['carat'], hist=False)
```

```python
sns.scatterplot(data=df, x='carat', y='price')
plt.show()
```

```python
for cut in df['cut'].drop_duplicates().tolist():
    sns.scatterplot(data=df[df['cut'] == cut], x='carat', y='price')
    plt.title(cut)
    plt.show()
```

```python
sns.violinplot(data=df, x='clarity', y='price_log1p')
```

```python
sns.violinplot(data=df, x='clarity', y='carat')
```

```python
for cut in df['cut'].drop_duplicates().tolist():
    sns.scatterplot(data=df[df['cut'] == cut], x='carat', y='price', hue='clarity')
    plt.title(cut)
    plt.show()
```
